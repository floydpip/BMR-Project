#### Render video in real time


```python
def start_video():
    # Connect to webcam
    cap = cv2.VideoCapture(0)

    # loop through each frame of live video until closed
    while cap.isOpened():
        ret, frame = cap.read()

        # Show image
        cv2.imshow('Webcam', frame)

        # Press q to close
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

```

```python
start_video()
```
