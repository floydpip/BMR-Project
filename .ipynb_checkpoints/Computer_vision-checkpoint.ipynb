## 1. Connect to webcam


```python
# Connect to webcam
cap = cv2.VideoCapture(1)

# Get a frame from the capture device
ret, frame = cap.read()

# True if connected to camera
print(ret)
```

```python
plt.imshow(frame)
```

```python
def take_photo():
    cap = cv2.VideoCapture(1)
    ret, frame = cap.read()
    cv2.imwrite('images/arena.jpg', frame)
    cap.release()
```

```python
take_photo()
```
